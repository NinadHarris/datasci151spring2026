Lecture 12: Zip 
When using a list of numbers in a print statement use str(...) to turn it into a string
zip: for x, y in zip(x_title, y_title)
Lets you loop through multiple lists (even more than two lists) at the same time by pairing them together



Zip with list of two:

````python
movie_titles = ["The Umbrellas of Cherbourg", "M3gan", "Psycho"]
movie_ratings = [7.8, 6.3, 8.5]

for title, rating in zip(movie_titles, movie_ratings):
    print("The movie " + title + " has an IMDB rating of " + str(rating))
````

Zip with NumPy arrays and tuples:

````python
title_list = ["The Umbrellas of Cherbourg", "M3gan", "Psycho"]

import numpy as np 
rating_array = np.array([7.8, 6.3, 8.5])

year_tuple = ("1964", "2022", "1960")

director_list = ["Jacques Demy", "Gerard Johnstone", "Alfred Hitchcock"]

for title, rating, year, director in zip(title_list, rating_array, year_tuple, director_list):
    message = "'" + title + "' (" + year + ") was directed by " + director + ". IMDB rating: " + str(rating)
    print(message)
````

Lecture 14: Functions
A function is a block of reusable code to perform a specific task. We can use functions to ignore repitition. 

For median:

````python
import numpy as np
np.random.seed(158354304)
array_1 = np.random.normal(size = 100)
array_2 = np.random.normal(size = 97)
array_3 = np.random.normal(size = 120)

median_1 = np.median(array_1)
print("Median of first array:", median_1)

median_2 = np.median(array_2)
print("Median of second array:", median_2)

median_3 = np.median(array_3)
print("Median of third array:", median_3)
````

Built in functions: print(), type(), round(), abs(), len(), str(), divmod()-- quotient and gives remainder

````python
# Arguments: 2 numbers

# Return value: A tuple containing the quotient and remainder

dividend = 32
divisor = 5

div_tuple = divmod(dividend, divisor)

print(div_tuple)
print(type(div_tuple))

quotient = div_tuple[0]
remainder = div_tuple[1]

print("Quotient:", quotient)
print("Remainder:", remainder)
````

````python
top_number = 32
bottom_number = 5

q, r = divmod(top_number, bottom_number)

print(type(q))
print(type(r))

print("Quotient:", q)
print("Remainder:", r)
````

Using "_" is a way to ignore an output of a function is you're only interested in one

Can also make custom functions ex: def quadratic_formula(a,b,c):

````python
def quadratic_formula(a, b, c): 
    # just copy and paste code from the cell above
    discrim = b**2 - 4*a*c
    x_plus = -b + np.sqrt(discrim)
    x_plus = x_plus/(2*a)

    x_minus = -b - np.sqrt(discrim)
    x_minus = x_minus/(2*a)

    # return two values as output
    return x_plus, x_minus
output = quadratic_formula(5, 15, 10)

print(type(output))
print(output[0])
print(output[1])
    
````

In general functions take the form:

```python

    #---- DEFINE
    def my_function(parameter):
        body
        return expression

    #---- RUN
    my_function(parameter = argument) 

    #---- RUN
    my_function(argument)
```

For multiple arguments or multiple return values:
```python

    #---- DEFINE
    def my_function(parameter1, parameter2, parameter3):
        body
        return expression1, expression2

    #---- RUN
    my_function(parameter1 = arg1, parameter2 = arg2, parameter3 = arg3) 

    #---- RUN
    my_function(arg1, arg2, arg3)
```
